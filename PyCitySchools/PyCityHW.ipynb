```python
# Dependencies
import pandas as pd
import numpy as np
import os


# define file path
school_file = os.path.join("Resources/schools_complete.csv")
student_file = os.path.join ("Resources/students_complete.csv")

# read files
school_df = pd.read_csv(school_file)
student_df = pd.read_csv(student_file)

# merge files into one.
merge_df = school_df.merge(student_df, how = 'left', on = 'school_name')






```

```python
# Calculation of unique schools
unique_schools = merge_df['school_name'].unique()
school_count = len(unique_schools)

# Calculation of total students
total_students = merge_df['student_name'].count()

# Calculation of Total Budget
total_budget = merge_df['budget'].sum()

# Calculation of average math and reading score
avg_math = merge_df['math_score'].mean()
avg_reading = merge_df['reading_score'].mean()

# Calculation of students passing math and reading
pass_math = merge_df.loc[merge_df['math_score'] >= 70]['math_score'].count()
percent_pass_math = pass_math/total_students

pass_reading = merge_df[merge_df['reading_score'] >= 70]['reading_score'].count()
percent_pass_reading = pass_reading/total_students

# Calculation of Students passing math and reading
overall_passing = merge_df.loc[(merge_df['math_score'] >=70) & (merge_df['reading_score'] >= 70)]['student_name'].count()/total_students

#Produce the table
district_summary = pd.DataFrame({
    "Total Schools": [school_count],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_reading],
    "% Passing Math": [percent_pass_math],
    "% Passing Reading": [percent_pass_reading],
    "% Passing Overall": [overall_passing]
})
display(district_summary)


```

```python
# Group by school
school_names = merge_df.set_index('school_name').groupby(['school_name'])

# School types
schooltypes = school_df.set_index('school_name')['type']

# Total students per school
totalstudents = school_names['Student ID'].count()

# Total school budget
school_budget = school_df.set_index("school_name")['budget']

# Total student budget
student_budget = school_df.set_index('school_name')['budget']/school_df.set_index('school_name')['size']

# Average scores by school
average_math = school_names['math_score'].mean()
average_reading = school_names["reading_score"].mean()

# Percent Passing Scores
passing_math = merge_df[merge_df['math_score'] >= 70].groupby('school_name')['Student ID'].count()/totalstudents
passing_reading = merge_df[merge_df["reading_score"] >= 70].groupby('school_name')['Student ID'].count()/totalstudents
passing_overall = merge_df[(merge_df['math_score'] >= 70) & (merge_df['reading_score'] >= 70)].groupby('school_name')['Student ID'].count()/totalstudents

#Produce the table
school_summary = pd.DataFrame({
    "School Type": schooltypes,
    "Total Students": totalstudents,
    "School Budget": school_budget,
    "Per Student Budget": student_budget,
    "Average Math Score": average_math,
    "Average Reading Score": average_reading,
    "% Passing Math": passing_math,
    "% Passing Reading": passing_reading,
    "% Passing overall": passing_overall

})
display(school_summary)
```

```python
# values sorted by overall passing rate, displaying top 5
top_5 = school_summary.sort_values("% Passing overall", ascending= False)
top_5.head()
```

```python
# values still sorted by overall passing rate, however, displaying bottom 5
bottom_5 = top_5.tail()
bottom_5 = bottom_5.sort_values('% Passing overall')
bottom_5.head()
```

```python
#Math Scores by grade
ninth_math = merge_df.loc[merge_df["grade"] == '9th'].groupby('school_name')['math_score'].mean()
tenth_math = merge_df.loc[merge_df["grade"] == '10th'].groupby('school_name')['math_score'].mean()
eleventh_math = merge_df.loc[merge_df["grade"] == '11th'].groupby('school_name')['math_score'].mean()
twelfth_math = merge_df.loc[merge_df["grade"] == '12th'].groupby('school_name')['math_score'].mean()

#Produce the table
math_by_grade = pd.DataFrame({
    '9th': ninth_math,
    '10th': tenth_math,
    '11th': eleventh_math,
    '12th': twelfth_math
})
display(math_by_grade)
```

```python
#Reading Scores by grade
ninth_reading = merge_df.loc[merge_df["grade"] == '9th'].groupby('school_name')['reading_score'].mean()
tenth_reading = merge_df.loc[merge_df["grade"] == '10th'].groupby('school_name')['reading_score'].mean()
eleventh_reading = merge_df.loc[merge_df["grade"] == '11th'].groupby('school_name')['reading_score'].mean()
twelfth_reading = merge_df.loc[merge_df["grade"] == '12th'].groupby('school_name')['reading_score'].mean()

#Produce the table
reading_by_grade = pd.DataFrame({
    '9th': ninth_reading,
    '10th': tenth_reading,
    '11th': eleventh_reading,
    '12th': twelfth_reading
})
display(reading_by_grade)
```

```python
#Determined the bin sizes
bins = [0, 584.999, 614.999, 644.999, 999999]
group_name = ['< $585', "$585 - 614", "$615 - 644", "> $644"]
merge_df['school_spending_bins'] = pd.cut(merge_df['budget']/merge_df['size'], bins, labels = group_name)

#grouping by spend
spending = merge_df.groupby('school_spending_bins')

#calcs
ave_math = spending['math_score'].mean()
ave_reading = spending['reading_score'].mean()
math_pass = merge_df[merge_df['math_score'] >=70].groupby('school_spending_bins')['Student ID'].count()/spending['Student ID'].count()
reading_pass = merge_df[merge_df['reading_score'] >=70].groupby('school_spending_bins')['Student ID'].count()/spending['Student ID'].count()
overall = merge_df[(merge_df['math_score'] >=70) & (merge_df['reading_score'] >= 70)].groupby('school_spending_bins')['Student ID'].count()/spending['Student ID'].count()

#Produce the table
scores_by_spending = pd.DataFrame({
    "Average Math Score": ave_math,
    "Average Reading Score": ave_reading,
    "% Passing Math": math_pass,
    "% Passing Reading": reading_pass,
    "% Overall Passing": overall
})
display(scores_by_spending)
```

```python
#Create bins again, this time for size of school
bins = [0,999,1999,9999999]
group_name = ['Small(<1000)','Medium(1000-2000)','Large(2000-5000)']
merge_df['school_size_bins'] = pd.cut(merge_df['size'], bins, labels = group_name)

#grouping by size
size = merge_df.groupby('school_size_bins')

#calcs
ave_math = size['math_score'].mean()
ave_reading = size['reading_score'].mean()
math_pass = merge_df[merge_df['math_score'] >=70].groupby('school_size_bins')['Student ID'].count()/size['Student ID'].count()
reading_pass = merge_df[merge_df['reading_score'] >=70].groupby('school_size_bins')['Student ID'].count()/size['Student ID'].count()
overall = merge_df[(merge_df['math_score'] >=70) & (merge_df['reading_score'] >= 70)].groupby('school_size_bins')['Student ID'].count()/size['Student ID'].count()

#Produce the table
scores_by_size = pd.DataFrame({
    "Average Math Score": ave_math,
    "Average Reading Score": ave_reading,
    '% Passing Math': math_pass,
    '% Passing Reading': reading_pass,
    "Overall Passing Rate": overall
            
})
display(scores_by_size)
```
